```python
import pandas as pd
python = {"RollNum":[22,852,33,44,55,66,77,999,88,99], "python":[78,85,89,87,67,87,90,76,98,100]}
python = pd.DataFrame(python)


numpy = {"RollNum":[22,33,44,55,66,77,11,88,99,333], "numpy":[67,45,89,87,67,87,90,76,98,100]}
numpy = pd.DataFrame(numpy)
```

```python
innerMerge = pd.merge(python, numpy)
```

```python
innerMerge
```

```python
outerMerge = pd.merge(python, numpy, how='outer')
```

```python
outerMerge
```

```python
leftMerge = pd.merge(python, numpy, how='left')
```

```python
leftMerge
```

```python
rightMerge = pd.merge(python, numpy, how='right')
```

```python
rightMerge
```
