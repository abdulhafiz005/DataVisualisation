```python
import pandas as pd
import numpy as np
students = {'name':["Ali","Faraz","Kamran","Faizan","Nasir","Asad","Danish"],
            'Age':[20,21,24,21,20,23,np.nan],
            'Python':[89,49,78,96,100,78, 80],
            'Numpy':[78,87,np.nan,89,75,95,85],
            'Pandas':[40,75,85,48,60,89, 96]}

data = pd.DataFrame(students)
data
```

```python
data.fillna({'Age': data['Age'].mean(), "Python":0, "Numpy":0, "Pandas":0}, inplace=True)
```

```python
data
```

```python
data['Total'] = data['Python'] + data['Numpy'] + data['Pandas']
data
```

```python
data['Percentage'] = round(data['Total']/300*100, 2)
data
```

```python
data["Result"] = ["Pass" if m >= 150 else "Fail" for m in data["Total"]]
```

```python
data
```

```python
data["Grade"] = ["F" if g < 40 else ("D" if g < 50 else ("C" if g < 60 else ("B" if g < 70 else "A"))) for g in data["Percentage"]]
data
```

```python
data['Result2'] = ["Pass" if py >= 50 and np >= 50 and pd >= 50 else "Fail" for py, np, pd in zip(data['Python'], data['Numpy'], data['Pandas'])]
data
```

```python
data = data.drop(['Result'], axis=1)
```
